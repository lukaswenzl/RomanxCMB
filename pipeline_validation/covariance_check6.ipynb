```python

filename = "../6x2pt_Roman_SO.fits"
hdul = fits.open(filename)
print(hdul.info())
cov_cosmosis = hdul[1].data
```

```python
#cosmosis covariance, assuming fsky 0.05 throughout
filename = "../6x2pt_Roman_SO_cosmolike.fits"
hdul = fits.open(filename)
print(hdul.info())
cov_cosmolike = hdul[1].data
```

```python

filename = "../6x2pt_Roman_SO_type2.fits"
hdul = fits.open(filename)
print(hdul.info())
cov_cosmosis_type2 = hdul[1].data
```

```python
x = np.arange(len(np.diagonal(cov_cosmosis)))
plt.figure(figsize=(20,5))
plt.title("cosmolike comparison")
plt.plot(x, np.log10(np.diagonal(cov_cosmosis)), label="cosmosis")


plt.text(0, -15, "shear ->")
plt.axvline(x=1100)
plt.text(1100, -11, "galaxy x shear ->")
plt.axvline(x=1740)
plt.text(1740, -11, "shear x cmb")
plt.axvline(x=1940)
plt.text(1940, -20, "galaxy")
plt.axvline(x=2140)
plt.text(2140, -11, "galaxy x cmb")
plt.axvline(x=2340)
plt.text(2340, -11, "cmb")


plt.plot(x, np.log10(np.diagonal(cov_cosmolike)), "--",label="cosmolike")
# plt.plot(x, np.diagonal(cov_log), "--",label="all 0.05")
# plt.plot(x, np.diagonal(cov_log_cmb0_4), "--",label="cmb 0.4, rest 0.05")

plt.legend()
plt.ylabel("Diagonal log Cov")
plt.xlabel("Index of covariance")

plt.figure(figsize=(20,5))
plt.title("ratio")
plt.plot(x,np.diagonal(cov_cosmolike)/np.diagonal(cov_cosmosis) );
tmp = np.diagonal(cov_cosmolike)/np.diagonal(cov_cosmosis)
for i in range(len(tmp)):
    if(not (i % 20)):
        tmp[i]=1
plt.plot(x, tmp,"--", label="ignore first bin")

plt.plot(x,np.diagonal(cov_cosmolike)/np.diagonal(cov_cosmosis_type2) );


#plt.plot(x,np.diagonal(cov_cosmolike_ng)/np.diagonal(cov_cosmosis) );

#plt.plot(x,np.diagonal(cov_more_ell)/np.diagonal(cov) );
plt.text(0, 1.3, "shear ->")
plt.axvline(x=1100)
plt.text(1100, 1.3, "galaxy x shear ->")
plt.axvline(x=1740)
plt.text(1740, 1.3, "shear x cmb")
plt.axvline(x=1940)
plt.text(1940, 1.3, "galaxy")
plt.axvline(x=2140)
plt.text(2140, 1.3, "galaxy x cmb")
plt.axvline(x=2340)
plt.text(2340, 1.3, "cmb")

mask1 = (np.diagonal(cov_cosmolike)/np.diagonal(cov_cosmosis) < 1.1)
mask2 = (np.diagonal(cov_cosmolike)/np.diagonal(cov_cosmosis) > 0.90)
inrange = np.sum(np.logical_and(mask1, mask2))
print(inrange / len(mask1))

mask1 = (tmp < 1.1)
mask2 = (tmp > 0.90)
inrange = np.sum(np.logical_and(mask1, mask2))
print(inrange / len(mask1))
#93% of diagonals match within 10%. ignoring the first bin its 98%
```
