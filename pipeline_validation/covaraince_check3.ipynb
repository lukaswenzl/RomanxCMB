```python
#direct comparison of the diagonal
x = np.arange(len(np.diagonal(cov)))
#diff = np.diagonal(cov_log) - np.diagonal(cosmolike_gaussian_log)
plt.figure(figsize=(15,4))
plt.plot(x, np.diagonal(cov_log), label="cosmosis")
plt.plot(x, np.diagonal(cosmolike_gaussian_log), "--",label="cosmolike")
plt.legend()
plt.ylabel("diagonal log Cov")
```
